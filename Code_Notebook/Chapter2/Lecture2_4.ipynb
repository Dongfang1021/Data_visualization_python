## 2.4 柱状图

```python
import pandas as pd
import numpy as np
from matplotlib.ticker import MultipleLocator
import matplotlib as mpl
mpl.rcParams["font.family"] = 'Arial'  #默认字体类型
mpl.rcParams["mathtext.fontset"] = 'cm' #数学文字字体
import matplotlib.pyplot as plt
```

### 参数介绍

## Error bar

```python
fig=plt.figure()
ax=plt.gca()
err_kw={'capsize':5, 'capthick':2,'ecolor':'r'}
ax.bar(labels, men_means, width, yerr=men_std, color='lightblue', label='Men',error_kw=err_kw)

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')

# savefig
plt.tight_layout()
plt.savefig('../../figures/Chapter2/Lecture2_4_basic_2.pdf')
plt.show()
```

## 水平柱状图

```python
def autolabel(rects):
    for rect in rects:
        x=rect.get_width()
        y=rect.get_y()+rect.get_height()/2
        ax.text(x*1.01, y,str(x),ha='left',va='center')
        
fig=plt.figure()
ax=plt.gca()
bar_men=ax.barh(labels, men_means, width, color='lightblue', label='Men',ec='k',lw=0.2)
autolabel(bar_men)

ax.set_xlabel('Scores')
ax.set_title('Scores by group and gender')

# savefig
plt.tight_layout()
plt.savefig('../../figures/Chapter2/Lecture2_4_basic_3.pdf')
plt.show()
```

## 填充图案

```python
from matplotlib.patches import Ellipse, Polygon

fig=plt.figure()
ax=plt.gca()
bar_men=ax.bar(labels, men_means, width, color='None', label='Men',ec='k',lw=0.9) #hatch='*'
patterns = ('---', '+++', 'x', '\\\\', '*', 'o', 'O', '.')
for bar, pattern in zip(bar_men,patterns):
    bar.set_hatch(pattern)
mpl.rcParams['hatch.linewidth'] = 0.3 

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')

# savefig
plt.tight_layout()
plt.savefig('../../figures/Chapter2/Lecture2_4_basic_4.pdf')
plt.show()
```
